- A variable can be a number, a character, a string (sequence of character), a collection of numbers, etc. 
- The equal sign (=) is used to assign a value to a variable (again, be careful to not confuse with testing for equality ==).

## Advanced (optional)
Plot temperature data from NCEP reanalysis data. xarray.open_dataset can open and decode a dataset from a file path or an OpenDAP URL.

The dataset that we will be using is here:  <br>https://www.esrl.noaa.gov/psd/thredds/catalog/Datasets/ncep.reanalysis/surface/catalog.html

```python
import xarray as xr
file = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis/surface/air.sig995.2023.nc'
ds = xr.open_dataset(file)
ds
```

Plot temperature for the last time step.

```python
ds.air.isel(time=-1).plot()
```

```python
# Make a prettier plot
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
lev1=np.arange(-50,36,1)
proj = ccrs.PlateCarree()     # map projection

fig = plt.figure(figsize=(9,5))
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
(ds.air.isel(time=-1)-273.15).plot.contourf(ax=ax, transform=ccrs.PlateCarree(),levels=lev1)
ax.coastlines()
```

```python
ds.time.isel(time=0)
```

```python
# Plot global T for January 1st 00 UTC
ds.air.sel(time='2023-01-01T00:00:00.000000000').plot()
```
